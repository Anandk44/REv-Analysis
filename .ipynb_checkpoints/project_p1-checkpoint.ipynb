```python
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
from pyspark.sql.functions import *


df = spark.read.csv("C:\\Users\\ANAND\\Desktop\\p1code\\PROJECT2-main\\DATASETS\\loan.csv", inferSchema = True, header = True)
df1 = spark.read.csv("C:\\Users\\ANAND\\Desktop\\p1code\\PROJECT2-main\\DATASETS\\credit card.csv")
def mainmenu():
    print('Welcome to Rev Data Analysis')
    choice = 0
    while True :
       
        print()
        print('1.press 1 to view all data in dataset ')
        print('2.press 2 to print number of coloumns  ')
        print('3. press 3 to view a specific customer  ')
        print('4. press 4 to view loans in each category  ')
        print('5. press 5 to view the information of loan customers ')
        print('6. press 6 to view the income of customers')
        print('7. press 7 to view who are not eligible for loan ')
        print("8. press 8 to view list of potential customers  ")
        print("9 . press 9 to view the list of defaulters ")
        print('9. Exit')
        print()
        
        choice = int(input('select your option  : '))
        
        if choice == 1:
            showalldata()
        elif choice == 2:
            userchoicerows()
        elif choice == 3:
            viewspecificcustomer()
        elif choice == 4:
            loansineachcategory() 
        elif choice == 5:
            loantracker()
        elif choice == 6:
            checkincome()
        elif choice == 7:
            loaneligibility()
        elif choice == 8 :
            listofpotentialcustomers()
        elif choice == 9 :
            listofpotentialdefaulters()
        elif choice == 10:
            print("Thank you for using Rev analysis ")
            break
        else:
            print('Invalid Option !!!')
            print('Please Select A Valid Option')
            print()
            input('Press Enter To Continue...')
def showalldata():
              df.show()
def userchoicerows() :
              inp = int(input("Enter number of rows you want to view "))
              df.show(inp)
def loansineachcategory() :
              df.groupBy("Loan Category").count().orderBy("count", ascending = False).show()

def viewspecificcustomer() :
             inp = input("Enter customer id to view customer ")
             df.select("Customer_ID", "Age", "Occupation", "Marital Status", "Income").filter(df.Customer_ID==inp).show()

def loantracker() :
        while True :
             inp = int(input("enter choice  to check the loans of customer as per criteria  "))
             
             if inp == 1 :
                    df.select("Customer_ID", "Age", "Occupation", "Income").filter(df["Loan Amount"]>="1,00,000").show(100)
                    
             elif inp == 2 :
                df.select("Customer_ID", "Age", "Occupation",  "Income",).filter(df["Loan Amount"]>="2,00,000").show(50)
                break
             elif inp ==  3 :
                df.select("Customer_ID", "Age", "Occupation",  "Income").filter(df["Loan Amount"]>"3,00,000").show(70)
                break
             elif inp == 4 :
                df.select("Customer_ID", "Age", "Occupation",  "Income",).filter(df["Loan Amount"]>"4,00,000").show(100)
                break
             elif inp == 5 :
                df.select("Customer_ID", "Age", "Occupation",  "Income").filter(df["Loan Amount"]>"5,00,000").show(100)
                break
             elif inp == 6 :
                df.select("Customer_ID", "Age", "Occupation",  "Income").filter(df["Loan Amount"]>"6,00,000" ).show(50)
                break
             elif inp == 7 :
                df.select("Customer_ID", "Age", "Occupation",  "Income").filter(df["Loan Amount"]>"7,00,000").show(50)
                break
             else :
                break
def checkincome() :
            while True :
                  print()
                  print('1.press 1 to check income of customer between 0 to 50000 ')
                  print('2.press 2 to check income of customer between 50000 to 100000  ')
                  print('3. press 3 to check income of customer more than 100000 ')
                  ch = int(input("Enter choice : "))
                  if ch == 1 :
                        df.select("Customer_ID" , "occupation","Income").filter(df["Income"]<"50000").show(100)
                        break
                  elif ch == 2 :
                        df.select("Customer_ID" , "occupation",'Income').filter(df["Income"]>"50000").show(100)
                        break
                  elif ch == 3 :
                        df.select("Customer_ID" , "occupation","Income").filter(df["Income"]>"100000").show(100)
                        break
                  elif ch >= 4 :
                        print("please choose correct option ")
                        break
def loaneligibility() :
    print("Non Eligible Customers are : ")
    print()
    df.select("Customer_ID","Occupation","Family Size","Income","Expenditure","Overdue").filter((df["Expenditure"]>df["Income"])).show()
    
def listofpotentialcustomers() :
    print("List of potential non defaulter customers ")
    print()
    df.select("Customer_ID","Occupation","Income","Expenditure","Overdue").filter((df["Income"]> df["Expenditure"]) & (df["Income"] > '50000') & (df["Overdue"]<=5)).show(50)
    
def listofpotentialdefaulters() :
    print("List of potential defaulters : ")
    df.select("Customer_ID" ,"Occupation" ,"Loan Amount").filter((df[" Returned Cheque"]>"1") & (df["Overdue"]>"1") & (df["Expenditure"] > df["Income"])).show()

mainmenu()
```
